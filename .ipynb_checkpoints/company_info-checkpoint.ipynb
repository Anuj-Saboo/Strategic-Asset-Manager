```python
import pandas as pd
from sqlalchemy import create_engine
from yahoofinancials import YahooFinancials
from datetime import date
engine = create_engine('postgresql://postgres:postgres@localhost:5432/stocks')
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
data_100 = pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')
table = data[0]
table_100 = data_100[2]
sp_list = table[['Symbol','Security','GICS Sector','GICS Sub Industry','Headquarters Location','CIK']]
sp_list = sp_list.rename(columns={'Symbol': 'symbol', 'Security': 'company_name','GICS Sector': 'sector','GICS Sub Industry':'subsector','Headquarters Location':'hq','CIK' : 'cik'})
```

```python
import json
engine = create_engine('postgresql://postgres:postgres@localhost:5432/stocks')
df = pd.read_sql_query('select symbol,company_name,sector,subsector,hq,cik from "company_info_tree"',con=engine)

df_final = pd.DataFrame()
#symbols = ['BRK.B']
symbols = table_100['Symbol'].tolist()
#symbols = df['symbol'].tolist()

yahoo_financials_tech = YahooFinancials(symbols)

daily_stocks_prices = yahoo_financials_tech.get_historical_price_data('2019-12-31','2020-01-01', 'daily')
for symbol in symbols:
    try:
        prices_list = daily_stocks_prices.get(symbol).get("prices")
        df_prices = pd.DataFrame.from_dict(prices_list)
        df_prices = df_prices[['high','low','open','close','volume','formatted_date']]
        subsetDataFrame = df[df['symbol'] == symbol]
        company_name = subsetDataFrame['company_name']
        df_prices['year'] = pd.to_datetime(df_prices['formatted_date']).dt.year
        df_prices['company_name'] = company_name.values[0]
        df_prices['symbol'] = symbol
        df_final = df_final.append(df_prices,ignore_index=True)
        df_prices.to_sql('stocks_prices', con=engine,if_exists='append',index=False)
    except:
        print(symbol)
        pass
```

```python
result = pd.merge(df_final,df,on='symbol',how='inner')
result = result[['cik','symbol','formatted_date','company_name_x','sector','subsector','high','low','open','close','volume','year']]
result = result.rename(columns={'company_name_x': 'company'})
result.head()
```

```python
result['mean'] = (result['high'] + result['low'] + result['open'] + result['close']) / 4
result.head()
```

```python
result.to_csv('treemap_1daymean.csv')
```
