```python
import wikipedia
import pandas as pd

table_100 = pd.read_csv('sp100.csv')

symbol_list = []
companyname_list = []
companydata_list = []

for index, row in table_100.iterrows():
    company_name = row['Name']
    symbol_list.append(row['Symbol'])
    companyname_list.append(company_name)
    
    orig = wikipedia.summary(company_name, sentences=4)
    replace_1 = orig.replace('== History ==','')
    replace_2 = replace_1.replace('\n','')
    companydata_list.append(replace_2)
    
companydata = pd.DataFrame(list(zip(symbol_list,companyname_list, companydata_list)), 
               columns =['Symbol','Name', 'Data']) 
```

```python
companydata.to_csv('company_data.csv',index=False)
```
